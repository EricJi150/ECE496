```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df1 = pd.read_csv('4_Channel_Model_Eval.csv')
matrix_data1 = df1.to_numpy() * 100  # Multiply all values by 100

df2 = pd.read_csv('5_Channel_Model_Eval.csv')
matrix_data2 = df2.to_numpy() * 100  # Multiply all values by 100

matrix_data = matrix_data2 - matrix_data1

# Formatting the data to two decimal places for annotations
formatted_matrix_data = np.array([[f"{value:.2f}" for value in row] for row in matrix_data])

row_labels = ['ProGAN', 'StyleGAN', 'ProjectedGAN', 'Diff-StyleGAN2', 'Diff-ProjectedGAN', 'DDPM', 'IDDPM', 'ADM', 'PNDM', 'LDM']
column_labels = ['ProGAN', 'StyleGAN', 'ProjectedGAN', 'Diff-StyleGAN2', 'Diff-ProjectedGAN', 'DDPM', 'IDDPM', 'ADM', 'PNDM', 'LDM']

# Create a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(matrix_data, annot=formatted_matrix_data, fmt="s", cmap='viridis', xticklabels=column_labels, yticklabels=row_labels)

# Customize the plot
plt.title('5-Channel Model minus 4-Channel Model Accuracy', size=24)
plt.xlabel('Trained On', size=18)
plt.ylabel('Tested On', size=18)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and display the plot
plt.savefig('Model_Difference.png', facecolor='white')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Confusion matrix data
# df = [[99932/1100, 68/1100], [6213/1100, 3787/1100]] #4 channel
df = [[99827/1100, 173/1100], [5999/1100, 4001/1100]] #5 channel

row_labels = ['Fake', 'Real']
column_labels = ['Fake', 'Real']

# Convert to numpy array and format the matrix data as strings
formatted_matrix_data = np.array([[f"{value:.2f}" for value in row] for row in df])

# Create a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(np.array(df), annot=formatted_matrix_data, fmt="s", cmap='viridis', xticklabels=column_labels, yticklabels=row_labels)

# Customize the plot
plt.title('5-Channel Multiclass Model Real vs Fake Confusion Matrix', size=24)
plt.ylabel('True label', size=18)
plt.xlabel('Predicted label', size=18)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and display the plot
plt.savefig('5-Channel-Real-Fake-Model.png', facecolor='white')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = pd.read_csv('5_Channel_Multi_Model_Eval.csv')

row_labels = ['ADM', 'DDPM', 'Diff-ProjectedGAN', 'Diff-StyleGAN2', 'IDDPM', 'LDM', 'PNDM', 'ProGAN', 'ProjectedGAN', 'StyleGAN', 'Real']
column_labels = ['ADM', 'DDPM', 'Diff-ProjectedGAN', 'Diff-StyleGAN2', 'IDDPM', 'LDM', 'PNDM', 'ProGAN', 'ProjectedGAN', 'StyleGAN', 'Real']

formatted_matrix_data = np.array([[f"{value:.0f}" for value in row] for row in df.values])

# Create a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(df, annot=formatted_matrix_data, fmt="s", cmap='viridis', xticklabels=column_labels, yticklabels=row_labels)

# Customize the plot
plt.title('5-Channel Multiclass Model Confusion Matrix', size=24)
plt.ylabel('True label', size=18)
plt.xlabel('Predicted label', size=18)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and display the plot
plt.savefig('5-Channel-Multi-Model.png', facecolor='white')
plt.show()
```
